# Arithmetic Script

### 1. Addition

```python
# Defining Variables
x = 2
y = 3

addition = x + y

print('Sum: {}'.format(addition))
```

### 2. Subtraction

```python
subtraction = y - x

print('Difference: {}'.format(subtraction))
```

### 3. Multiplication

```python
multiplication = x * y

print('Product: {}'.format(multiplication))
```

### 4. Division

```python
division = x / y

print('Quotient: {}'.format(division))
```

### 5. Formula

```python
formula_01 = (x + y) + 1

print('Formula 01 Result: {}'.format(formula_01))
```
